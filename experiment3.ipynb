```python
from pathlib import Path
import sys

sys.path.append(str(Path(".").resolve()))

from Core.Model import LoadFullModel, LoadLiteModel, ComputeIOUs
from Core.ImageHandling import LoadPILImages
import numpy as np

from itertools import chain

liteModel = LoadFullModel(Path(r"SURE\datasets\handpicked\2\models\100\UNet_100_images_E063"))
image_path = Path(r"SURE\datasets\handpicked\2\subsets\100")

folders = [
    "testing",
    "training",
    "validation"
]

def get_images_from_context(c: str):
    images = [
        LoadPILImages(image_path / x / c) for x in folders 
    ]
    return list(chain.from_iterable(images))

images = get_images_from_context("images")
segmentations = get_images_from_context("segmentations")

ious = ComputeIOUs(liteModel, images, segmentations)

ious
```
